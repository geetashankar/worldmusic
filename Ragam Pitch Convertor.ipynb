```python
import mingus.core.notes as notes
```

```python
swaram_to_int_dict = {
    "S": 0,
    "R1": 1,
    "R2": 2, 
    "R3": 3,
    "G1": 2,
    "G2": 3,
    "G3": 4,
    "M1": 5,
    "M2": 6,
    "P": 7,
    "D1": 8,
    "D2": 9,
    "D3": 10,
    "N1": 9,
    "N2": 10,
    "N3": 11,
}
```

```python
def check_valid_swaram(notes):
    for i in notes:
        if i not in swaram_to_int_dict.keys():
            print(i, " is an invalid swaram.")
            return
    print("All swarams are valid.")
```

```python
def swaram_to_int_converter(curr_swaram):
    for swaram, num in swaram_to_int_dict.items():
        if swaram == curr_swaram:
            return num
```

```python
def swaram_to_pitch(swaram,shruthi):
    swaram_int = swaram_to_int_converter(swaram)
    shruthi_int = notes.note_to_int(shruthi)
    return notes.int_to_note((swaram_int + shruthi_int)%12)
```

```python
def main():
    shruthi = input("Enter a shruthi: ")
    shruthi = shruthi.upper()
    arohanam = input("Enter the arohanam with a space in between each note: ")
    arohanam = arohanam.upper().split()
    check_valid_swaram(arohanam)
    avarohanam = input("Enter the avarohanam with a space in between each note: ")
    avarohanam = avarohanam.upper().split()
    check_valid_swaram(avarohanam)
    arohanam_pitches = []
    for swaram in arohanam:
        arohanam_pitches.append(swaram_to_pitch(swaram, shruthi))
    avarohanam_pitches = []
    print("The arohanam is ",arohanam_pitches)
    for swaram in avarohanam:
        avarohanam_pitches.append(swaram_to_pitch(swaram, shruthi))
    print("The avarohanam is ",avarohanam_pitches)
```

```python
main()
```

```python
import pysynth_b as psb # a, b, e, and s variants available

''' (note, duration)
Note name (a to g), then optionally a '#' for sharp or
'b' for flat, then optionally the octave (defaults to 4).
An asterisk at the end means to play the note a little 
louder.  Duration: 4 is a quarter note, -4 is a dotted 
quarter note, etc.'''
song = (
  ('a3', 4), ('a#3', 4), ('c#4', 4), 
  ('d4', 4),  ('e4', 4), ('f4', 4),
  ('g#4', 4), ('a4', 4), ('a4', 4), ('g#4', 4), ('f4', 4), 
  ('e4', 4),  ('d4', 4), ('c#4', 4),
  ('a#3', 4), ('a3', 4),
)

# Beats per minute (bpm) is really quarters per minute here
psb.make_wav(song, fn = "song.wav", leg_stac = .7, bpm = 135)
```
